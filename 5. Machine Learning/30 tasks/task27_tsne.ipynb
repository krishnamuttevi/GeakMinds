t-distributed Stochastic Neighbor Embedding

It is a powerful dimensionality reduction technique used to visualize high-dimensional data in 2D or 3D, Can capture non linear complex relationships

eg:

Handwritten digits (64 features from 8Ã—8 pixels)

Word embeddings (300-dimensional vectors)

Gene expression data

```python
from sklearn import datasets
import matplotlib.pyplot as plt

digits = datasets.load_digits()

# Plot first 4 digit images with their labels
_, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, label in zip(axes, digits.images, digits.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.set_title(f"Label: {label}")
```

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt


X = digits.data
y = digits.target

tsne = TSNE(n_components=2, random_state=42)
X_2d = tsne.fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X_2d[:,0], X_2d[:,1], c=y, cmap='tab10', alpha=0.7)
plt.colorbar(label='Digit Label')
plt.title("t-SNE Visualization of Digits")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Needed for 3D plotting

# Load data
digits = load_digits()
X = digits.data
y = digits.target

# Apply t-SNE with 3 components
tsne = TSNE(n_components=3, random_state=42)
X_3d = tsne.fit_transform(X)

# Plotting in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_3d[:, 0], X_3d[:, 1], X_3d[:, 2], c=y, cmap='tab10', alpha=0.7)

# Add colorbar and labels
legend1 = ax.legend(*scatter.legend_elements(), title="Digits", loc="upper right")
ax.add_artist(legend1)

ax.set_title("3D t-SNE Visualization of Digits")
ax.set_xlabel("t-SNE Component 1")
ax.set_ylabel("t-SNE Component 2")
ax.set_zlabel("t-SNE Component 3")

plt.show()
```
