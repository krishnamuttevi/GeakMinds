```python
df_test['Age'] = df_test['Age'].fillna(df_test['Age'].mode()[0])
df_test.drop(columns='Cabin')
df_test['tot_family']=df_test['SibSp']+df_test['Parch']+1
df_test['IsAlone'] = (df_test['tot_family'] == 1).astype(int)
df_test['AgeBin'] = pd.cut(df_test['Age'], bins=[0, 12, 18, 35, 60, 100], labels=['Child', 'Teen', 'YoungAdult', 'Adult', 'Senior'])
df_test['FareBin'] = pd.qcut(df_test['Fare'], q=4, labels=['Low', 'Mid', 'High', 'VeryHigh'])
df_test = pd.get_dummies(df_test, columns=['Sex'], drop_first=True)
df_test = pd.get_dummies(df_test, columns=['Embarked','AgeBin','FareBin'])
df_test=df_test.drop(columns=['Name','Ticket','Cabin'])
```

```python
from sklearn.linear_model import LogisticRegression

model_logi = LogisticRegression(max_iter=1000)  # max_iter is often increased to ensure convergence
model_logi.fit(df_train.drop(columns='Survived'), df_train['Survived'])
```

```python
from sklearn.ensemble import RandomForestClassifier

model_Random = RandomForestClassifier()
model_Random.fit(df_train.drop(columns='Survived'), df_train['Survived'])
```

```python
df_test["Fare"]=df_test["Fare"].fillna(7.8292)
```

```python
logi_preds = model_logi.predict(df_test.drop(columns='Survived'))
rf_preds = model_Random.predict(df_test.drop(columns='Survived'))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score

def evaluate_model(name, y_true, y_pred, model, X_val):
    print(f"\n{name} Performance:")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall:", recall_score(y_true, y_pred))
    print("ROC-AUC:", roc_auc_score(y_true, model.predict_proba(X_val)[:, 1]))

evaluate_model("Logistic Regression", df_test['Survived'], logi_preds, model_logi, df_test.drop(columns='Survived'))
evaluate_model("Random Forest",df_test['Survived'], rf_preds, model_Random, df_test.drop(columns='Survived'))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Logistic Regression
cm_logi = confusion_matrix(df_test['Survived'], logi_preds)
disp_logi = ConfusionMatrixDisplay(confusion_matrix=cm_logi, display_labels=[0, 1])
disp_logi.plot()
plt.title("Logistic Regression Confusion Matrix")
plt.show()

# Random Forest
cm_rf = confusion_matrix(df_test['Survived'], rf_preds)
disp_rf = ConfusionMatrixDisplay(confusion_matrix=cm_rf, display_labels=[0, 1])
disp_rf.plot()
plt.title("Random Forest Confusion Matrix")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import roc_curve, auc

logi_proba = model_logi.predict_proba(df_test.drop(columns='Survived'))[:, 1]

fpr, tpr, thresholds = roc_curve(df_test['Survived'], logi_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], '--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()

print(f'AUC Score: {roc_auc:.4f}')

j_scores = tpr - fpr
optimal_idx = np.argmax(j_scores)
optimal_threshold = thresholds[optimal_idx]

print(f'Optimal Threshold: {optimal_threshold:.4f}')
print(f'TPR at optimal: {tpr[optimal_idx]:.4f}')
print(f'FPR at optimal: {fpr[optimal_idx]:.4f}')
```
