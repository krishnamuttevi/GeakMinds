Implement matrix multiplication manually (without using np.dot or @).

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

if A.shape[1]==B.shape[0]:
  C=np.zeros((A.shape[0],B.shape[1]))
  for i in range(A.shape[0]):
    for j in range(B.shape[1]):
      for l in range(B.shape[0]):
        C[i][l]+=A[i][l]*B[l][j]
  print(C)

else:
  print('not possible')
```

Create a 3D array of shape (3, 4, 5) filled with random integers. Normalize the array values to the range 0-1.

```python
arr=np.random.randint(1,100,60).reshape(3,4,5).astype('float')
mea=np.mean(arr)
stda=np.std(arr)
for i in np.ndindex(arr.shape):
  arr[i]=(arr[i]-mea)/stda
arr
```

```python
arr=np.random.randint(1,100,60).reshape(3,4,5).astype('float')
mini=np.min(arr)
maxi=np.max(arr)
if mini!=maxi:
  for i in np.ndindex(arr.shape):
    arr[i]=(arr[i]-mini)/(maxi-mini)
arr
```

```python
def inverse(arr):
  deter=np.linalg.det(arr)
  if deter!=0:
    print(f" since determinent is {deter} inverse exists")
    a_inv=np.linalg.inv(arr)

    return a_inv
  else:
    print(f" since determinent is {deter} inverse does not exists")

arr=np.random.randint(1,101,size=16).reshape(4,4)
print('inverse is : \n',inverse(arr))
```

```python
import numpy as np

def simulate_markov_chain(P, initial_state, n_steps):

    n_states = P.shape[0]
    states = np.zeros(n_steps + 1, dtype=int)
    states[0] = initial_state

    current = initial_state
    for t in range(n_steps):
        current = np.random.choice(n_states, p=P[current])
        states[t + 1] = current

    return states

P = np.array([
        [0.7, 0.2, 0.1],
        [0.3, 0.4, 0.3],
        [0.2, 0.3, 0.5]
])

print("Transition Matrix:")
print(P)
states = simulate_markov_chain(P, 0, 10)
print(f"states: {states}")
print(f"State frequencies: {np.bincount(states) / len(states)}")
```
