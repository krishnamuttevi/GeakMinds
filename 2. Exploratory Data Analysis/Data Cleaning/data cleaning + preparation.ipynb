# **Section 2: Handling Missing Data**

Check for missing values in a dataset and print the count of missing values per column.

```python
emp_df.isnull().sum()
```

Remove all rows containing missing values from the dataset

```python
emp_df=emp_df.dropna()
emp_df
```

Impute missing numerical values using the mean of the column.

```python
mean=emp_df['Salary'].mean()
emp_df['Salary']=emp_df['Salary'].fillna(mean)
emp_df
```

```python
mode=emp_df['Department'].mode()[0]
emp_df['Department']=emp_df['Department'].fillna(mode)
emp_df
```

Fill missing values using forward fill (propagating last valid observation forward).

```python
emp_df['Department']=emp_df['Department'].fillna(method="ffill")
emp_df
```

```python
def min_max(df,col):
  min=df[col].min()
  max=df[col].max()
  df[col]=(df[col]-min)/(max-min)
  return df
emp_df=min_max(emp_df,'Salary')
emp_df
```

```python
def z_score(df,col):
  mean=df[col].mean()
  std=df[col].std()
  df[col]=(df[col]-mean)/std
  return df
emp_df=z_score(emp_df,'Age')
emp_df
```

Apply logarithmic transformation to reduce skewness in a dataset

```python
# skewness - measure of assymetry
c=1
bef_skew=emp_df['Salary'].skew()
k=np.log(emp_df['Age']+c)
print("before skew",bef_skew)
print("after skew",k.skew())
emp_df['Age']=k
emp_df.head()
```

```python
#sqt - only + ve data needs shilft for -ve data
bef_skew=emp_df['Salary'].skew()
if emp_df['Salary'].min()<0:
  shf=emp_df['Salary'].min()*-1
  emp_df['Salary']=emp_df['Salary']+shf
k=np.sqrt(emp_df['Salary']+c)
k.skew()
print("before skew",bef_skew)
print("after skew",k.skew())
emp_df['Salary']=k
emp_df.head()
```

```python
# box cox strictly +ve for Stabilize variance ,Make data more normally distributed
def boxcox(x, lmbda):
    if x.min() <= 0:
        x = x + abs(x.min()) + 1
    if lmbda == 0:
        return np.log(x)
    else:
        return (x**lmbda - 1) / lmbda
c = 0.3
emp_df['Salary'] = boxcox(emp_df['Salary'], c)
emp_df.head()
```

# class for the entire assignment

```python
class Assignment3:
  def __init__(self,file_path):
    self.file_path=file_path
    self.df=pd.read_csv(file_path)
    print(self.df.head())
  def remove_duplicates(self):
    self.df=self.df.drop_duplicates()
    return self.df
  def conv(self,col,dtype):
    self.df[col]=self.df[col].astype(dtype)
    return self.df
  def iqr(self,col):
    q1=np.percentile(self.df[col],25)
    q3=np.percentile(self.df[col],75)
    iqr=q3-q1
    l_limit=q1-1.5*iqr
    u_limit=q3+1.5*iqr
    indi=self.df[(self.df[col]<l_limit) | (self.df[col]>u_limit)].index
    print("Outliers : ",self.df[col][indi])
    self.df=self.df.drop(indi)
    return self.df
  def loweer(self,col):
    self.df[col]=self.df[col].apply(lambda x: x.lower() if isinstance(x,str) else x)
    return self.df
  def invalid_neg(self,col):
    indi=self.df[self.df[col]<0].index
    print("Invalid : ",self.df[col][indi])
    self.df=self.df.drop(indi)
    return self.df
  def ct_missing(self):
    print(self.df.isnull().sum())
  def drop_missing(self):
    self.df=self.df.dropna()
    return self.df
  def fill_missing(self,col,value):
    if value=="mean":
      value=self.df[col].mean()
    elif value=="mode":
      value=self.df[col].mode()[0]
    self.df[col]=self.df[col].fillna(value)
    return self.df
  def fill_missing_ffill(self,col):
    self.df[col]=self.df[col].fillna(method="ffill")
    return self.df
  def min_max(self,col):
    min=self.df[col].min()
    max=self.df[col].max()
    self.df[col]=(self.df[col]-min)/(max-min)
    return self.df
  def z_score(self,col):
    mean=self.df[col].mean()
    std=self.df[col].std()
    self.df[col]=(self.df[col]-mean)/std
    return self.df
  def log_transform(self,col,cons):
    self.df[col]=np.log(self.df[col]+cons)
    return self.df
  def sqt_smooth(self,col):
    if self.df[col].min()<0:
      shf=self.df[col].min()*-1
      self.df[col]=self.df[col]+shf
    self.df[col]=np.sqrt(self.df[col])
    return self.df
  def boxcox(self,col,cons):
    if self.df[col].min()<0:
      shf=self.df[col].min()*-1
      self.df[col]=self.df[col]+shf
    if cons==0:
      self.df[col]=np.log(self.df[col])
    else:
      self.df[col]=(self.df[col]**cons-1)/cons
    return self.df
  def desc(self):
    print(self.df.describe())
  def hist(self):
    self.df.hist()
  def boxplot(self):
    self.df.boxplot()
  def scatter(self,x,y):
    plt.scatter(self.df[x],self.df[y])
  def corr(self):
    cols=[]
    for i in self.df.columns:
      if self.df[i].dtype=="int":
        cols.append(i)
    corr=self.df[cols].corr()
    sns.heatmap(corr,annot=True)
```
