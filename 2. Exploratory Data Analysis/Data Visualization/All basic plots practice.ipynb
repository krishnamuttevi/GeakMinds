```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C', 'A'],
    'Value': np.random.randint(10, 100, 10),
    'Date': pd.to_datetime(['2023-01-15', '2023-01-20', '2023-02-01', '2023-02-10',
                            '2023-03-01', '2023-03-05', '2023-04-01', '2023-04-10',
                            '2023-05-01', '2023-05-15'])
}
df = pd.DataFrame(data)
df = df.sort_values("Date")

df['Change'] = df.groupby('Category')['Value'].diff().fillna(0)  # one date nxt to value
df['Color'] = df['Change'].apply(lambda x: 'green' if x >= 0 else 'red')

plt.figure(figsize=(10, 6))
plt.bar(df['Date'].dt.strftime('%Y-%m-%d'), df['Change'], color=df['Color'])

plt.title("Diverging Bar Chart: Change in Values Over Time")
plt.xlabel("Date")
plt.ylabel("Change")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

np.random.seed(42)
days = 100
dates = range(days)
initial_price = 100

prices = [initial_price]
for i in range(1, days):
    change = np.random.normal(0, 2)  # Random daily change
    new_price = prices[-1] + change
    prices.append(max(50, min(200, new_price)))  # Keep price between 50-200

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, days)
ax.set_ylim(min(prices) - 10, max(prices) + 10)
ax.set_xlabel('Days')
ax.set_ylabel('Stock Price ($)')
ax.set_title('Animated Stock Price Movement')
ax.grid(True, alpha=0.3)


line, = ax.plot([], [], 'b-', linewidth=2)
price_text = ax.text(0.02, 0.95, '', transform=ax.transAxes, fontsize=12,
                     bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.8))

def animate(frame):

    x_data = dates[:frame+1]
    y_data = prices[:frame+1]

    line.set_data(x_data, y_data)

    if frame < len(prices):
        price_text.set_text(f'Current Price: ${prices[frame]:.2f}')

    return line, price_text

anim = animation.FuncAnimation(fig, animate, frames=days, interval=100,
                              blit=True, repeat=True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats


np.random.seed(42)
group1 = np.random.normal(50, 10, 1000)    # mean=50, std=10
group2 = np.random.normal(60, 15, 1000)
group3 = np.random.normal(55, 8, 1000)
group4 = np.random.normal(65, 12, 1000)

plt.figure(figsize=(12, 8))

plt.hist(group1, bins=40, alpha=0.7, label='Group 1', color='red', density=True)
plt.hist(group2, bins=40, alpha=0.6, label='Group 2', color='blue', density=True)
plt.hist(group3, bins=40, alpha=0.5, label='Group 3', color='green', density=True)
plt.hist(group4, bins=40, alpha=0.4, label='Group 4', color='orange', density=True)

x_range = np.linspace(10, 100, 200)

kde1 = stats.gaussian_kde(group1)
kde2 = stats.gaussian_kde(group2)
kde3 = stats.gaussian_kde(group3)
kde4 = stats.gaussian_kde(group4)

plt.plot(x_range, kde1(x_range), color='red', linewidth=3, alpha=0.8)
plt.plot(x_range, kde2(x_range), color='blue', linewidth=3, alpha=0.8)
plt.plot(x_range, kde3(x_range), color='green', linewidth=3, alpha=0.8)
plt.plot(x_range, kde4(x_range), color='orange', linewidth=3, alpha=0.8)

plt.xlabel('Values', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.title('Multi-Layered Density Plot with Different Transparency Levels', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)

stats_text = f"""
Group 1 Mean: {np.mean(group1):.1f}
Group 2 Mean: {np.mean(group2):.1f}
Group 3 Mean: {np.mean(group3):.1f}
Group 4 Mean: {np.mean(group4):.1f}
"""
plt.text(0.02, 0.98, stats_text, transform=plt.gca().transAxes,
         verticalalignment='top', fontsize=10,
         bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

plt.tight_layout()
plt.show()
```
