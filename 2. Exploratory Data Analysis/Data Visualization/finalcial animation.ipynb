```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

np.random.seed(42)
days = 100
prices = [100]
for i in range(1, days):
    prices.append(np.clip(prices[-1] + np.random.normal(0, 2), 50, 200))

fig, ax = plt.subplots(figsize=(10, 5))
line, = ax.plot([], [], 'b-', lw=2)
ax.set_xlim(0, days)
ax.set_ylim(min(prices)-10, max(prices)+10)
ax.set_xlabel('Day')
ax.set_ylabel('Price ($)')
ax.set_title('Live Stock Price Trend')
price_label = ax.text(0.02, 0.95, '', transform=ax.transAxes, fontsize=12,
                      bbox=dict(facecolor='yellow', alpha=0.8))

def animate(frame):
    line.set_data(range(frame+1), prices[:frame+1])
    price_label.set_text(f'Price: ${prices[frame]:.2f}')
    return line, price_label

ani = animation.FuncAnimation(fig, animate, frames=days, interval=100, blit=True)
plt.tight_layout()
plt.show()
```
